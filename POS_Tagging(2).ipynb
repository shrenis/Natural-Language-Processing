```python
import nltk
from nltk.corpus import gutenberg
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
with open ("C:\\Users\\SHRENI\\Desktop\\VIT\\3rd_sem\\Dataset\\Treasure_Island.txt" , encoding="utf8") as f:
    island = f.read()
```

```python
print(island)
```

```python
tokenized_sentences = nltk.tokenize.sent_tokenize(island)
print(tokenized_sentences)
```

```python
sentences = []
for sentence in tokenized_sentences:
    sentences.append(sentence.rstrip('\r\n'))
print(sentences)
```

```python
print(f"Total number of sentences in the corpus: {len(sentences)}")
```

```python
tokenized_words = []
for sent in sentences:
    words = nltk.wordpunct_tokenize(sent)
    for word in words:
        tokenized_words.append(word)

print(tokenized_words)
```

```python
print(f"Total number of words in the corpus: {len(tokenized_words)}")
```

```python
## stemming the corpus
import nltk
from nltk.stem import SnowballStemmer
snowStem = SnowballStemmer('english')
stem_words = []
for word in tokenized_words:
    stem_words.append(snowStem.stem(word))

print(stem_words)
```

```python
tagged_words = nltk.pos_tag(stem_words)
```

```python
print(f"Total number of tagged words in the corpus: {len(tagged_words)}")
```

```python
tagged_words_fd = nltk.FreqDist(word for (word, tag) in tagged_words)
most_common_words = tagged_words_fd.most_common()
common_words = []
for word, count in most_common_words:
    common_words.append(word)
    
freq_words_joined = ' '.join(common_words)
print(freq_words_joined)
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='Black', 
                stopwords = stopwords, 
                min_font_size = 10).generate(freq_words_joined) 
  
# plot the WordCloud image                        
plt.figure(figsize = (16, 16), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for (word, tag) in tagged_words)
common_len_words = []
for i in range(3, 20):
    if i in cfd.conditions():
        common_len_words.append(nltk.FreqDist(cfd[i]).max())

words_joined = ' '.join(common_len_words)
print(words_joined)
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate(words_joined) 
  
# plot the WordCloud image                        
plt.figure(figsize = (5, 5), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
